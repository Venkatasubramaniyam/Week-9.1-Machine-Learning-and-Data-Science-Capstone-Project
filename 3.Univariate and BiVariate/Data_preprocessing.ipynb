```python
import pandas as pd
```

```python
dataset=pd.read_csv("telecom_churn.csv")
```

```python
dataset
```

```python
def quanqual(dataset):
    quan=[]
    qual=[]
    for columnname in dataset.columns:
        if (dataset[columnname].dtypes=='str'):
            qual.append(columnname)
        else:
            quan.append(columnname)
    return quan,qual
```

```python
quan,qual=quanqual(dataset)
```

```python
dataset[quan]
```

```python
numu=dataset[quan]
```

```python
cate=dataset[qual]
```

```python
cate
```

```python
merg=[numu,cate]
```

```python
preprocssed=pd.concat(merg,axis=1)  #axis=1 column wise merger
```

```python
preprocssed
```

```python
preprocssed.isnull().sum()
```

```python
preprocssed.to_csv("preprocessed_data.csv",index=False)
```
